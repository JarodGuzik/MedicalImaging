```python
# set up the losses

class SSD:
    """
    SSD Loss for image registration
    """
    def loss(self, fixed, moving):
        dif = fixed - moving
        difSq = dif ** 2
        return difSq
    

# losses = [vxm.losses.NCC().loss, SSD().loss]
```

```python
# First run the model with the NCC loss
# model1.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=1e-4), loss=vxm.losses.NCC().loss)
model1.compile(optimizer='Adam', loss=vxm.losses.NCC().loss)

```

```python


def vxm_data_generator(x_data, batch_size=4):
    """
    Generator that takes in data of size [N, H, W, D], and yields data for
    our custom vxm model in sequential pairs instead of random sampling.

    inputs:  moving [bs, H, W, D], fixed image [bs, H, W, D]
    outputs: moved image [bs, H, W, D], zero-gradient [bs, H, W, 3]
    """

    # preliminary sizing
    N = x_data.shape[0]  # Number of images (N)
    vol_shape = x_data.shape[1:]  # extract data shape (H, W, D)
    ndims = len(vol_shape)

    # Prepare a zero array the size of the deformation
    zero_phi = np.zeros([batch_size, *vol_shape, ndims])

    # Generate all unique image pairs (i, j) where i != j
    pairs = [(i, j) for i in range(N) for j in range(N) if i != j]

    # Iterate through pairs in batches
    num_batches = len(pairs) // batch_size
    pair_idx = 0
    
    while True:
        # Select a batch of pairs
        batch_pairs = pairs[pair_idx:pair_idx + batch_size]

        # Prepare inputs for the batch
        moving_images = np.array([x_data[i, ..., np.newaxis] for i, j in batch_pairs])
        fixed_images = np.array([x_data[j, ..., np.newaxis] for i, j in batch_pairs])
        inputs = [moving_images, fixed_images]

        # Prepare outputs (the 'true' moved image, which is the fixed image)
        outputs = [fixed_images, zero_phi]

        yield (inputs, outputs)

        # Update the pair index for the next batch
        pair_idx = (pair_idx + batch_size) % len(pairs)
        if pair_idx == 0:
            # If we have iterated through all pairs, restart the loop
            pass
        
def get_batch_from_generator(generator, batch_number):
    for i, (inputs, outputs) in enumerate(generator):
        if i == batch_number:
            return inputs, outputs
    return None
```

```python
from sklearn.preprocessing import StandardScaler

ssdScaler = StandardScaler()

def ssdProcess(fullData):
    processed = fullData.reshape(fullData.shape[0], fullData.shape[1], -1)
    its = fullData.shape[0]
    
    for iii in range(its):
        processed[iii,:,:] = ssdScaler.fit_transform(processed[iii,:,:])
        
    processed = processed.reshape(fullData.shape)
    
    return processed

npTrainProc = ssdProcess(npTrain)
npTestProc = ssdProcess(npTest[0:11,:,:,:])


# foo = npTrain.reshape(npTrain.shape[0], npTrain.shape[1], -1)
# print(foo.shape)
        
```

## Task 1 training comments:
While NCC could train well with large batches, I often struggled with memory errors or kernel freezing when trying to run large batch size training. So while the models have the capability programmically to handle batch sizes larger than 1, I did online training to ensure that the training would actually finish in a timely manner. To help improve accuracy with the smaller batch size than the tutorial, I ran the model for more epochs and more iterations per epoch to compensate. The SSD loss model would not train well as well with large batch sizes, so I kept the batch size to 1. To get high results, SSD required some preprocess of the data to normalize it, while NCC could train on the data without normalization with the same result due to the error calculation of NCC having its own nromalization factor. Despite this, I trained the NCC loss model with the normalized data since the process needed to be done for the SSD loss model anyway and they may as well be given consistent training data.

# Task 3: Dice and Hausdorff Before and After Image Registration:
## Please note: Dice coefficient and Hausdorff distance calculations done here will ignore the 0 labels (regions that are background and not part of the brain).  
## Voxels have spacing of 1,1,1 so they are all evenly spaced in all dimensions. While no actual real distance is listed in the metadata, we assume it is in mm

```python
   

# from neurite.neurite.tf.metrics import Dice
# from voxelmorph.voxelmorph.py.utils import dice

# Need number of classes for diceCo
origMasks = npTest[11:, :, :, :]
all_classes = (np.unique(origMasks[0,:,:,:]))
nb_classes = len(all_classes)

# # calculate mean dice coefficient for moved masks
# def diceCo(yTrue, yPred):
#     dicetype = Dice(dice_type='hard', input_type='max_label', nb_labels=nb_classes)
#     dice_coef = dicetype.mean_dice(yTrue, yPred)
#     scalarDice = dice_coef.numpy()
#     return scalarDice

# def diceCo(yTrue, yPred, allClasses):
#     totalDice = dice(yTrue, yPred, allClasses) # standard dice from voxelmorph excludes the 0 label
#     # I have edited the dice calculation so that if a label is not in the region it shows up as -1 in the dice score
#     valid_dice_scores = totalDice[totalDice != -1] 
    
#     avgDice = np.mean(valid_dice_scores)
   
    
#     return avgDice

def diceCo(yTrue, yPred, allClasses):
    realClass = allClasses[allClasses != 0]
    classLen = len(realClass) # exclude the 0 class
    scores = np.zeros(classLen) - 1
    
    yTflat = yTrue.reshape(-1)
    yPflat = yPred.reshape(-1)
    
    
    for iji in range(classLen):
        cmult1 = yTflat == realClass[iji]
        cmult2 = yPflat == realClass[iji]
        
        cmult1 = cmult1.astype(int)
        cmult2 = cmult2.astype(int)
        
        tANDp = np.sum(cmult1*cmult2)
        tORp = np.sum(cmult1) + np.sum(cmult2)
        scores[iji] = 2 * tANDp / tORp
        
    return scores
        
    
# The distance is the same in all dimensions 
    
def HausDist(yTrue, yPred, nb_classes):
    # we will ignore the 0 class
    C = nb_classes - 1
    hausVec = np.zeros(C)-1
    for ii in range(C):
        oneHotTrue = yTrue == ii+1
        oneHotPred = yPred == ii+1
        
        oneHotTrue = oneHotTrue.astype(int)
        oneHotPred = oneHotPred.astype(int)
      
        
        hausD = hausdorff_distance(oneHotTrue, oneHotPred)
        
        pixel2dist = voxel_spacing[0]
        
        hausVec[ii] = hausD
        
        
        
    return hausVec



    
```

```python
testMask = npTest[11:, :, :, :]
testPicNorm = npTestProc[0:11, :, :, :]

test_generator = vxm_data_generator(testPicNorm, 1)
mask_generator = vxm_data_generator(testMask, 1)
totalBatches = 110 # 11*10 is 110 pairs

# set up vectors to store dice scores and hausdorff distance
C = nb_classes-1

# Dice and Haus for no registration
preRegDice = np.zeros(C)
preRegHaus = np.zeros(C)

# the two model losses
NCCDice = np.zeros(C)
NCCHaus = np.zeros(C)
SSDDice = np.zeros(C)
SSDHaus = np.zeros(C)

warp_model = networks.Transform(vol_size, interp_method='nearest')

# exclude the 0 class
realClass = all_classes[all_classes != 0]

for jiji in range(totalBatches):
    test_input, _ = get_batch_from_generator(test_generator, jiji)
    mask_inputs, _ = get_batch_from_generator(mask_generator, jiji)

    test_pred1 = model1foo.predict(test_input,)
    test_pred2 = model2foo.predict(test_input,)
    
    moved_pred1 = test_pred1[0].squeeze()
    moved_pred2 = test_pred2[0].squeeze()
    
    pred_warp1 = test_pred1[1]
    pred_warp2 = test_pred2[1]

    # extract input moving and fixed for registration and scoring
    movingMask = np.squeeze(mask_inputs[0])
    fixedMask = np.squeeze(mask_inputs[1])
    
    # for masks
    
    warped_seg1 = warp_model.predict([movingMask[np.newaxis,...,np.newaxis], pred_warp1])
    warped_seg2 = warp_model.predict([movingMask[np.newaxis,...,np.newaxis], pred_warp2])
    
    warped_seg1 = np.squeeze(warped_seg1)
    warped_seg2 = np.squeeze(warped_seg2)
    
    # Calculate scores without registration
    preDice = diceCo(fixedMask, movingMask, all_classes)
    preHaus = HausDist(fixedMask, movingMask, nb_classes)
    
    # NCC Loss model scores
    NCCDiceVec = diceCo(fixedMask, warped_seg1, all_classes)
    NCCHausVec = HausDist(fixedMask, warped_seg1, nb_classes)
    
    # SSD Loss model scores
    SSDDiceVec = diceCo(fixedMask, warped_seg2, all_classes)
    SSDHausVec = HausDist(fixedMask, warped_seg2, nb_classes)
    
    # save the scores to be averaged
    preRegDice = preRegDice + preDice
    preRegHaus = preRegHaus + preHaus
    
    NCCDice = NCCDice + NCCDiceVec
    NCCHaus = NCCHaus + NCCHausVec
    
    SSDDice = SSDDice + SSDDiceVec
    SSDHaus = SSDHaus + SSDHausVec
    

# Get the average dice score and Hausdorff distance for each class over all batches

preRegDice = preRegDice / totalBatches
preRegHaus = preRegHaus / totalBatches

NCCDice = NCCDice / totalBatches
NCCHaus = NCCHaus / totalBatches

SSDDice = SSDDice / totalBatches
SSDHaus = SSDHaus / totalBatches

print('The average class Dice scores before registration are ', preRegDice)
print('The average class Haus distances before registration are ', preRegHaus, ' mm')

print('The average NCC model class Dice scores after registration are ', NCCDice)
print('The average NCC model class Haus distances after registration are ', NCCHaus, ' mm')

print('The average SSD model class Dice scores after registration are ', SSDDice)
print('The average SSD model class Haus distances after registration are ', SSDHaus, ' mm')
```

# Task 3 Comments:
Below are the printed out vectors for the dice scores and hausdorff distances for each class for the pre registration case as well as for each model. The dice scores are higher for pretty much all the classes after registration than before, showing that the model has improved the overlap of the moving image onto the fixed image. This also aligns with what was seen in the images shown in task 2 before. Oddly, the Hausdorff distances for the NCC loss model are actually larger, showing that while the overall registration is good (high dice scores) either the edges of the extreme regions are misaligning or there are outlier points in the classes being registered nowhere near where they should be. Meanwhile, the dice scores for the SSD loss model were better than the unregistered yet worse than the NCC loss model, but the Hausdorff distances were around the same or slightly lower than the unregistered hausdorff distances, meaning that the SSD loss model was able to improve registration without causing strange outlier abnormalities.
