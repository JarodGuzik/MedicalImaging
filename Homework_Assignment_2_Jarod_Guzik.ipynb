```python
# HW2 
# Import packages
import numpy as np
import os
import torch
from torchvision import models
from torchvision import datasets
from torch.utils.data import DataLoader, random_split
import torchvision.transforms as transforms
# from sklearn.model_selection import test_train_split as split
import SimpleITK as sitk
# from PIL import Image
# from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay
# import cv2 as cv
# from sklearn.preprocessing import MinMaxScaler
import matplotlib.pyplot as plt
```

```python
# Resize the images
# CovidPath = 'COVID/COVID/images'
# OpacPath = 'Lung_Opacity/Lung_Opacity/images'
# NormPath = 'Normal/Normal/images'
# ViralPath = 'Viral_Pneumonia/Viral_Pnuemonia/images'

dataPath = 'LungImages'

transform = transforms.Compose([
    transforms.Resize((224, 224)),  # Resize to 224x224
    transforms.ToTensor(),           # Convert image to tensor
    transforms.Normalize(            # Normalize with mean and std for VGG16
        mean=[0.485, 0.456, 0.406],
        std=[0.229, 0.224, 0.225]
    ),
])

data = datasets.ImageFolder(root=dataPath, transform=transform)

# # Map all the integer targets to instead be strings of their label name
# labelMap = {
#     0: 'Covid',
#     1: 'Lung_Opacity',
#     2: 'Normal',
#     3: 'Viral_Pneumonia'
# }

# # Change the targets to string names
# newTarget = [labelMap[target] for target in data.targets]

# # Update the dataset targets
# data.targets = newTarget
    
# Calculate the sizes of each split
train_size = int(0.75 * len(data))
val_size = int(0.1 * len(data))
test_size = len(data) - train_size - val_size

# Split the dataset
train_data, val_data, test_data = random_split(data, [train_size, val_size, test_size])

train_loader = DataLoader(train_data, batch_size=32, shuffle=True)
val_loader = DataLoader(val_data, batch_size=32, shuffle=False)
test_loader = DataLoader(test_data, batch_size=32, shuffle=False)

    
# set the cuda core as the device to use the GPU computation 
device = torch.device("cuda" if torch.cuda.is_available() else "cpu") 
```

# For this data set, the labels are as follows:
## 0 - Covid  
## 1 - Lung Opacity  
## 2 - Normal  
## 3 - Viral Pneumonia

## Train the VGG16 Model

```python
# Choose the vgg16 model

# device = torch.device("cpu")
modelVGG16 = models.vgg16(weights='DEFAULT')
# model.to(device) 

for param in modelVGG16.features.parameters():
    param.requires_grad = False
    
# print(model)

num_classes = 4  # 4 different classes
# the 7th layer in classifier is the fully connected layer that we need to train
modelVGG16.classifier[6] = torch.nn.Linear(in_features=4096, out_features=num_classes)

# put the computation on the gpu
modelVGG16.to(device) 

```

```python
# Set the model to training mode
modelVGG16.train()

# Define loss function and optimizer
criterion = torch.nn.CrossEntropyLoss()
optimizer = torch.optim.Adam(filter(lambda p: p.requires_grad, modelVGG16.parameters()), lr=0.001)

# Training loop
num_epochs = 15
for epoch in range(num_epochs):
    running_loss = 0.0
    
    for images, labels in train_loader:
        # Move data to the device if using GPU
        images, labels = images.to(device), labels.to(device)
        
        # print(f'Images device: {images.device}, Labels device: {labels.device}')
#         for param in model.parameters():
            
#             print(f'Model param device: {param.device}')

        
        
        # Zero the parameter gradients
        optimizer.zero_grad()
        
        # Forward pass
        outputs = modelVGG16(images)
        
        # Compute loss
        loss = criterion(outputs, labels)
        
        # Backward pass
        loss.backward()
        
        # Optimize
        optimizer.step()
        
        running_loss += loss.item()
    
    # Print statistics
    print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {running_loss/len(train_loader):.4f}')
    

    # Set the model to evaluation mode
    modelVGG16.eval()
    val_loss = 0.0
    correct = 0
    total = 0
    
    with torch.no_grad():
        for images, labels in val_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = modelVGG16(images)
            loss = criterion(outputs, labels)
            val_loss += loss.item()
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
    
    print(f'Validation Loss: {val_loss/len(val_loader):.4f}, Accuracy: {100 * correct / total:.2f}%')

# Save the trained model if needed
# torch.save(modelVGG16.state_dict(), 'vgg16_finetuned.pth')
```

```python
# Test for VGG16
for param in modelVGG16.parameters():
    param.requires_grad = False

modelVGG16.eval()

predsVec = []
labelVec = []

with torch.no_grad():  # Disable gradient calculation
    for images, labels in test_loader:
        images = images.to(device)  # Move images to the appropriate device
        labels = labels.to(device)  # Move labels to the appropriate device
        
        outputs = modelVGG16(images)  # Get model predictions
        _, predicted = torch.max(outputs, 1)  # Get the predicted class
        
        predsVec.extend(predicted.cpu().numpy())  # Store predictions
        labelVec.extend(labels.cpu().numpy())  # Store true labels
        
# Convert to numpy arrays
predsVec = np.array(predsVec)
labelVec = np.array(labelVec)
        
# Compute confusion matrix
cm = confusion_matrix(labelVec, predsVec)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)

# Compute accuracy score
accuracy = accuracy_score(labelVec, predsVec)
print(f'Test Accuracy for VGG16 is: {accuracy:.2f}')

# Plot confusion matrix
disp.plot(cmap=plt.cm.Blues)
plt.title('Confusion Matrix for VGG16')
plt.show()
```

```python
# there are 4 different labels, we will use an inverse frequency as the weighting for the cross entropy
train_labels = []  # Populate this with your actual training labels
for _, labels in train_loader:
    train_labels.extend(labels.numpy())
train_labels = torch.tensor(train_labels)

classCount = torch.bincount(train_labels)
totalCount = train_labels.size(0)

classWeights = 1/(classCount/totalCount)
cweight = classWeights.to(device)
```

## Discussion on accuracy

From the results shown through accuracy scores in test, VGG16 with regular cross entropy loss appears to perform the best. However, this does not tell the entire story. The accuracy of VGG16 is higher than the other models currently; however, the ResNet, GoogLeNet, and DenseNet models may perform better with very high numbers of epochs. Since I have only a single GPU core to work with, it is difficult to test large numbers of epochs in a timely manner (since we can only check out 8hr sessions for hipergator sessions). Also, even with only 15 epochs, we get relatively high accuracies with a medium training length (~ 2 - 2.5 hours for the 12 different models). While weighted cross entropy loss decreases the overall accuracy in test, by looking at the confusion matrices, we can see that the weighted cross entropy loss misclassifies the minority classes (Viral Pneumonia and Covid) at a lower rate than with regular cross entropy loss, although the improvement was much higher for Pneumonia (the most minority class) than for Covid. Even though more Normal and Lung Opacity classes are now classified as Covid or Pneumonia, we could consider these false positives which are much preferred to the false negtives of classifying Covid and Pneumonia as Lung Opacity and Normal (which we were getting many of with regular cross entropy loss). As for the Focal Loss, performance is highly dependent on the tuning of gamma. With a lower gamma, you see little to no improvement in actually classifying the minority classes, but with a high gamma, we overcorrect and get many false positives. In the paper Focal Loss for Dense Object Detection, they propose using alpha balancing (which I will show at the end of this HW) to help improve accuracy in focal loss. The paper gives gamma=2 ad alpha=0.25 as one properly matched combination of gamma and alpha, which I use for my own testing in this hw. To directly compare the alpha balanced focal loss with the regular focal loss, I keep gamma = 2 for all the focal loss tests.  

Next, I will see if data augmentation can improve the results, and I will use the VGG16 with weighted cross entropy loss to test. Even though it had 1% lower overall accuracy than VGG16 with regular cross entropy loss, the greatly improved accuracy on the minority classes makes it a good case to test data augmentation to see if we can now also reduce the false positives.

```python
# First create a function to add Gaussian noise to the images
class AddGaussianNoise(object):
    def __init__(self, mean=0.0, std=1.0):
        self.mean = mean
        self.std = std

    def __call__(self, tensor):
        # Ensure the input tensor is a float tensor
        if not tensor.dtype == torch.float32:
            raise ValueError("Input tensor must be a float tensor.")
        
        # Generate Gaussian noise
        noise = torch.randn_like(tensor) * self.std + self.mean
        
        # Add noise to the input tensor
        return tensor + noise

    
```

```python
# Select the DenseNet121 model

modelDense = models.densenet121(weights='DEFAULT')  

# freeze the parameters
for param in modelDense.parameters():
    param.requires_grad = False  # Freeze all layers before final layer


# Unfreeze the final layer (classifier)
num_features = modelDense.classifier.in_features
num_classes = 4  # 4 classes
modelDense.classifier = torch.nn.Linear(num_features, num_classes)

# device should still be set to the cuda core
modelDense.to(device)


# Train the DenseNet121 model

# Set the model to training mode
modelDense.train()

# Define gamma for focal loss
# Using gamma of 2 which was used in the focal loss for dense object detection paper
gamma = 2

# Alpha value from paper
alpha = 0.25

# Define optimizer
optimizer = torch.optim.Adam(filter(lambda p: p.requires_grad, modelDense.parameters()), lr=0.001)

# Training loop
num_epochs = 15
for epoch in range(num_epochs):
    running_loss = 0.0
    
    for images, labels in train_loader:
        # Move data to the device if using GPU
        images, labels = images.to(device), labels.to(device)      
        
        #### Section for Focal Loss ###
        
       # Zero the parameter gradients
        optimizer.zero_grad()
        
        # Forward pass
        outputs = modelDense(images)

        # Compute log softmax
        log_probs = torch.nn.functional.log_softmax(outputs, dim=1)

        # One-hot encoding of labels
        onehot = torch.zeros(labels.size(0), num_classes).to(device)
        onehot.scatter_(1, labels.unsqueeze(1), 1)  # One-hot encode the labels

        # Calculate probabilities
        probs = torch.exp(log_probs)

        # Compute focal loss
        loss = -onehot * (1 - probs) ** gamma * log_probs * alpha
        loss = loss.sum(dim=1)  # Sum across classes for each instance
        loss = loss.mean()  # Average over the batch
        
        #####
        
        # Backward pass
        loss.backward()
        
        # Optimize
        optimizer.step()
        
        running_loss += loss.item()
    
    # Print statistics
    print(f'Epoch [{epoch+1}/{num_epochs}], Loss: {running_loss/len(train_loader):.4f}')
    
    
    # Set the model to evaluation mode
    modelDense.eval()
    val_loss = 0.0
    correct = 0
    total = 0
    
    with torch.no_grad():
        for images, labels in val_loader:
            images, labels = images.to(device), labels.to(device)
            outputs = modelDense(images)
            
            
            #### Section for Focal Loss ###
            log_probs = torch.nn.functional.log_softmax(outputs, dim=1)

            # One-hot encoding of labels
            onehot = torch.zeros(labels.size(0), num_classes).to(device)
            onehot.scatter_(1, labels.unsqueeze(1), 1)  # One-hot encode the labels

            # Calculate probabilities
            probs = torch.exp(log_probs)

            # Compute focal loss
            loss = -onehot * (1 - probs) ** gamma * log_probs
            loss = loss.sum(dim=1)  # Sum across classes for each instance
            loss = loss.mean()  # Average over the batch
            ######
            
            
            val_loss += loss.item()
            _, predicted = torch.max(outputs.data, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()
    
    print(f'Validation Loss: {val_loss/len(val_loader):.4f}, Accuracy: {100 * correct / total:.2f}%')

# Save the trained model if needed
# torch.save(modelDense.state_dict(), 'denseNet121_finetuned.pth')

# Test for DenseNet121
for param in modelDense.parameters():
    param.requires_grad = False

modelDense.eval()

predsVec = []
labelVec = []

with torch.no_grad():  # Disable gradient calculation
    for images, labels in test_loader:
        images = images.to(device)  # Move images to the appropriate device
        labels = labels.to(device)  # Move labels to the appropriate device
        
        outputs = modelDense(images)  # Get model predictions
        _, predicted = torch.max(outputs, 1)  # Get the predicted class
        
        predsVec.extend(predicted.cpu().numpy())  # Store predictions
        labelVec.extend(labels.cpu().numpy())  # Store true labels
        
# Convert to numpy arrays
predsVec = np.array(predsVec)
labelVec = np.array(labelVec)
        
# Compute confusion matrix
cm = confusion_matrix(labelVec, predsVec)
disp = ConfusionMatrixDisplay(confusion_matrix=cm)

# Compute accuracy score
accuracy = accuracy_score(labelVec, predsVec)
print(f'Test Accuracy for DenseNet121 with Focal Loss is: {accuracy:.2f}')

# Plot confusion matrix
disp.plot(cmap=plt.cm.Blues)
plt.title('Confusion Matrix for DenseNet121 with Focal Loss')
plt.show()
```
